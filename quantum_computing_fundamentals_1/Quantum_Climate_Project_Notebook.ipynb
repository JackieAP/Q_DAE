## ⚖️ Classical vs Quantum Reasoning

### 1. State Representation
- Classical: One value at a time
- Quantum: Superposition (many states at once)

**Example:**
- Classical → One emission value
- Quantum → Encodes multiple years into qubits simultaneously

---

### 2. Probability
- Classical: Fixed probabilities
- Quantum: Probability amplitudes + interference

**Example:**
- Classical → Logistic regression predicts rise/fall
- Quantum → Uses probability from measurement of qubit

---

### 3. Information Flow
- Classical: Sequential
- Quantum: Entanglement (correlated information)

**Example:**
- Classical → Each year independent
- Quantum → Years entangled → captures relationships
