# Superposition 

Superposition is one state being with mulitple existing possibilities. 

Below is the representation of superposition in IBM quantum composer

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from numpy import pi

qreg_q = QuantumRegister(4, 'q')
creg_c = ClassicalRegister(4, 'c')
circuit = QuantumCircuit(qreg_q, creg_c)

circuit.h(qreg_q[0])
circuit.measure(qreg_q[0], creg_c[0])
circuit.draw()
```

# Entanglement 
Entanglement is a connection between two or more particles that link their properties together and act as a single unit. The state of one immediately influnces the state of being of the other, no matter how far apart they are.

Below is the representation of both particles in superpostion and the CNOT linking both.

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from numpy import pi

qreg_q = QuantumRegister(4, 'q')
creg_c = ClassicalRegister(4, 'c')
circuit = QuantumCircuit(qreg_q, creg_c)

circuit.h(qreg_q[0])
circuit.cx(qreg_q[0], qreg_q[1])
circuit.measure(qreg_q[0], creg_c[0])
circuit.measure(qreg_q[1], creg_c[1])
circuit.draw()
```

# The Deutsch's Algorithm 
This algorithm functions like a black box operation that has one input and one output but its different from classical computers because you cannot see how the "answer" is being calculated and its forced to double check while the algorithm uses quantum to check both inputs at once. The Deutsch algorithm calculates if the machine is giving a constant output, which means that not matter how many inputs you insert, it'll give you the same constant outputs. While the balanced gives you a different output each time you you insert an input

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from numpy import pi

qreg_q = QuantumRegister(4, 'q')
creg_c = ClassicalRegister(4, 'c')
circuit = QuantumCircuit(qreg_q, creg_c)

circuit.h(qreg_q[0])
circuit.x(qreg_q[1])
circuit.h(qreg_q[1])
circuit.cx(qreg_q[0], qreg_q[1])
circuit.h(qreg_q[0])
circuit.measure(qreg_q[0], creg_c[0])
circuit.draw()
```
